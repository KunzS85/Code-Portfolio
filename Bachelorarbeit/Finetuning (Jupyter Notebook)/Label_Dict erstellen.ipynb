```python
excel_path = '/content/drive/MyDrive/master_data.xlsx'
sheet_name = "snomed_master"
mdf = pd.read_excel(excel_path, sheet_name, index_col=None)
mdf = mdf.astype({'Code': str})
print(mdf)
print(len(mdf))
```

```python
label_dict = {}
count = 0
for l in df['Label'].unique():
  row = mdf.loc[mdf['Code'] == l]
  name = [x[1] for x in row.values]
  synonyms = [x[2] for x in row.values]
  if(l == "-1"): #or not name
    label_dict[l] = {
          'name': 'none of the trained concepts', 
          'synonym': [], 
          "code": l,
          'index': count}
    count += 1
  elif (not name):
    pass
  else:
    row = mdf.loc[mdf['Code'] == l]
    name = [x[1] for x in row.values]
    synonyms = [x[2] for x in row.values]
    label_dict[l] = {
      "name": name[0],
      "synonym": synonyms[0],
      "code": f"{l}",
      "index": count}
    count += 1
  

print(label_dict)
print(len(label_dict))
```
